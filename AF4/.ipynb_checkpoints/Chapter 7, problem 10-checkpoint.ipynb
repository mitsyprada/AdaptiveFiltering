```python
#unit variance signal
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mean_v = 0
std_v = 0.15
N = 150
a1 = 0.1
a2 = -0.8
v = np.random.normal(mean_v, std_v, size=N)#noise
u = np.zeros(N)
u[0]=6
for i in range(2,N,1):
    u[i] = -a1*u[i-1]-a2*u[i-2]+v[i]#AR process
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(v)
plt.title('Noise v(n) of variance '+format(std_v**2)+' and mean 0')
plt.subplot(212)
plt.plot(u)
plt.title('u(n) of variance '+format(np.std(u)**2)+' and mean '+format(np.mean(u)))
plt.tight_layout()
plt.show()
```

# a)

100 Monte Carlo simulations

```python
#100 Monte Carlo simulations
w=np.zeros((N+1,2))#2 tap weights
mu=0.2
delta=0.5
fMC=np.zeros((N,100))
f_=np.zeros(N)
fMC[0]=u[0]
w[0][0]=0
w[0][1]=0
for MC in range(100):
    v = np.random.normal(mean_v, std_v, size=N)#noise
    u = np.zeros(N)
    u[0]=10
    for i in range(2,N,1):
        u[i] = -a1*u[i-1]-a2*u[i-2]+v[i]#AR process
    for i in range(N):
        #predictor error
        fMC[i][MC]=u[i]-(np.dot(w[i][0],u[i-1])+np.dot(w[i][1],u[i-2]))
        #first tap weigh
        w[i+1][0]=w[i][0]+mu*u[i-1]*fMC[i][MC]/(np.absolute(u[i-1])**2+delta)
        #second tap weigh
        w[i+1][1]=w[i][1]+mu*u[i-2]*fMC[i][MC]/(np.absolute(u[i-2])**2+delta)

for i in range(N):
    for j in range(100):
        f_[i]+=fMC[i][j]   
    f_[i]=f_[i]/100
```

```python
plt.figure(figsize=(10,5))
#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f_**2,'r')
plt.legend(('mu=0.05'))
plt.xlabel('Number of adaptation cycles, n')
plt.ylabel('mean squared error')
plt.title('learning curve')
plt.show()
```

# b)

Power spectral plots of tap weight estimates

```python
e1=np.zeros(N)
e2=np.zeros(N)
for i in range(N):
    e1[i]=-a1-w[i][0]
    e2[i]=-a2-w[i][1]

plt.figure(figsize=(20,10))
plt.subplot(211)
plt.psd(e1)
plt.title('Power espectral plot of e1')
plt.subplot(212)
plt.psd(e2)
plt.title('Power espectral plot of e2')
plt.show()
```

# c)

Variating delta

```python
delta1=0.25
delta2=0.75
```

```python
#100 Monte Carlo simulations, delta1
w1=np.zeros((N+1,2))#2 tap weights
mu=0.2
fMC=np.zeros((N,100))
f1_=np.zeros(N)
fMC[0]=u[0]
w1[0][0]=0
w1[0][1]=0
for MC in range(100):
    v = np.random.normal(mean_v, std_v, size=N)#noise
    u = np.zeros(N)
    u[0]=10
    for i in range(2,N,1):
        u[i] = -a1*u[i-1]-a2*u[i-2]+v[i]#AR process
    for i in range(N):
        #predictor error
        fMC[i][MC]=u[i]-(np.dot(w1[i][0],u[i-1])+np.dot(w1[i][1],u[i-2]))
        #first tap weigh
        w1[i+1][0]=w1[i][0]+mu*u[i-1]*fMC[i][MC]/(np.absolute(u[i-1])**2+delta1)
        #second tap weigh
        w1[i+1][1]=w1[i][1]+mu*u[i-2]*fMC[i][MC]/(np.absolute(u[i-2])**2+delta1)

for i in range(N):
    for j in range(100):
        f1_[i]+=fMC[i][j]   
    f1_[i]=f1_[i]/100
```

```python
#100 Monte Carlo simulations, delta2
w2=np.zeros((N+1,2))#2 tap weights
mu=0.2
fMC=np.zeros((N,100))
f2_=np.zeros(N)
fMC[0]=u[0]
w2[0][0]=0
w2[0][1]=0
for MC in range(100):
    v = np.random.normal(mean_v, std_v, size=N)#noise
    u = np.zeros(N)
    u[0]=10
    for i in range(2,N,1):
        u[i] = -a1*u[i-1]-a2*u[i-2]+v[i]#AR process
    for i in range(N):
        #predictor error
        fMC[i][MC]=u[i]-(np.dot(w2[i][0],u[i-1])+np.dot(w2[i][1],u[i-2]))
        #first tap weigh
        w2[i+1][0]=w2[i][0]+mu*u[i-1]*fMC[i][MC]/(np.absolute(u[i-1])**2+delta2)
        #second tap weigh
        w2[i+1][1]=w2[i][1]+mu*u[i-2]*fMC[i][MC]/(np.absolute(u[i-2])**2+delta2)

for i in range(N):
    for j in range(100):
        f2_[i]+=fMC[i][j]   
    f2_[i]=f2_[i]/100
```

```python
plt.figure(figsize=(20,10))
plt.subplot(311)
#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f1_**2,'r')
plt.ylabel('mean squared error')
plt.title('learning curve. Delta=0.25')
plt.subplot(312)
#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f_**2,'r')
plt.ylabel('mean squared error')
plt.title('learning curve. Delta=0.5')
plt.subplot(313)
#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f2_**2,'r')
plt.xlabel('Number of adaptation cycles, n')
plt.ylabel('mean squared error')
plt.title('learning curve. Delta=0.75')
plt.show()
```

We can observe that the initial values of the mean square error are much smaller for earlier adaptation cycles when we increase delta. This is because using delta, gradient noise amplification can be diminished (this problem is worse for high values of |u(n)|), which diminishes the initial mean square error. For higher delta, the rate of convergence is faster. 

```python
e11=np.zeros(N)
e21=np.zeros(N)
for i in range(N):
    e11[i]=-a1-w1[i][0]
    e21[i]=-a2-w1[i][1]

e12=np.zeros(N)
e22=np.zeros(N)
for i in range(N):
    e12[i]=-a1-w2[i][0]
    e22[i]=-a2-w2[i][1]


plt.figure(figsize=(20,10))
plt.subplot(321)
plt.psd(e11)
plt.title('Power espectral plot of e1. Delta=0.25')
plt.subplot(322)
plt.psd(e21)
plt.title('Power espectral plot of e2. Delta=0.25')
plt.subplot(323)
plt.psd(e1)
plt.title('Power espectral plot of e1. Delta=0.5')
plt.subplot(324)
plt.psd(e2)
plt.title('Power espectral plot of e2. Delta=0.5')
plt.subplot(325)
plt.psd(e12)
plt.title('Power espectral plot of e1. Delta=0.75')
plt.subplot(326)
plt.psd(e22)
plt.title('Power espectral plot of e2. Delta=0.75')
plt.subplots_adjust(hspace=0.5)
plt.show()
```
