# Computer Experiment on Adaptive Prediction

The process is described by the following difference
equation:
u(n) = -au(n-1) + v(n)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mean_v = 0
std_v = 0.01
N = 500
a = -0.99
v = np.random.normal(mean_v, std_v, size=N)#noise
u = np.zeros(N)
u[0]=1
for i in range(1,N,1):
    u[i] = -a*u[i-1]+v[i]#AR process
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(v)
plt.title('Noise v(n) of variance '+format(std_v**2)+' and mean 0')
plt.subplot(212)
plt.plot(u)
plt.title('u(n) of variance '+format(np.std(u)**2)+' and mean '+format(np.mean(u)))
plt.tight_layout()
plt.show()
```

**Learning Curves for Varying mu**

```python
#mu=0.01
w=np.zeros(N+1)
f=np.zeros(N)
f[0]=u[0]
w[0]=0
mu=0.01
for i in range(N):
    f[i]=u[i]-w[i]*u[i-1]#predictor error
    w[i+1]=w[i]+mu*u[i-1]*f[i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
    #this is an auto regressive model, and the J(n) = f(n)^2

#mu=0.05
w1=np.zeros(N+1)
f1=np.zeros(N)
f1[0]=u[0]
w1[0]=0
mu1=0.05
for i in range(N):
    f1[i]=u[i]-w1[i]*u[i-1]#predictor error
    w1[i+1]=w1[i]+mu1*u[i-1]*f1[i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
    #this is an auto regressive model, and the J(n) = f(n)^2
    
#mu=0.1
w2=np.zeros(N+1)
f2=np.zeros(N)
f2[0]=u[0]
w2[0]=0
mu2=0.1
for i in range(N):
    f2[i]=u[i]-w2[i]*u[i-1]#predictor error
    w2[i+1]=w2[i]+mu2*u[i-1]*f2[i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
    #this is an auto regressive model, and the J(n) = f(n)^2

plt.figure(figsize=(20,10))
plt.plot(f**2,'r')#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f1**2,'b')
plt.plot(f2**2,'g')
plt.legend(('mu=0.01','mu=0.05','mu=0.2'),loc='upper right')
plt.xlabel('Number of adaptation cycles, n')
plt.ylabel('Mean square error')
plt.title('Learning curve')
plt.show()
```

The time constants of the learning curve and therefore the rate of convergence of the LMS algorithm are inversely proportional to the step-size parameter, mu.
Accordingly, keeping mu small results in a long rate of convergence and, therefore, a less statistically efficient LMS algorithm.
In other words, keeping mu small, robustness of the LMS algorithm (with a small misadjustment factor) is traded off for efficiency.

**The ensemble averaging was performed over 100 independent Monte Carlo runs of the experiment**

```python
mc_sim=100
w=np.zeros((mc_sim,N+1))
f=np.zeros((mc_sim,N))
w1=np.zeros((mc_sim,N+1))
f1=np.zeros((mc_sim,N))
w2=np.zeros((mc_sim,N+1))
f2=np.zeros((mc_sim,N))
for MC in range(mc_sim):#100 Monte Carlo simulations
    mean_v = 0
    std_v = 0.01
    N = 500
    a = -0.99
    v = np.random.normal(mean_v, std_v, size=N)#noise
    u = np.zeros(N)
    u[0]=1
    for i in range(1,N,1):
        u[i] = -a*u[i-1]+v[i]#AR process

    #mu=0.01

    f[MC][0]=u[0]
    w[MC][0]=0
    mu=0.01
    for i in range(N):
        f[MC][i]=u[i]-w[MC][i]*u[i-1]#predictor error
        w[MC][i+1]=w[MC][i]+mu*u[i-1]*f[MC][i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
        #this is an auto regressive model, and the J(n) = f(n)^2

    #mu=0.05
    f1[MC][0]=u[0]
    w1[MC][0]=0
    mu1=0.05
    for i in range(N):
        f1[MC][i]=u[i]-w1[MC][i]*u[i-1]#predictor error
        w1[MC][i+1]=w1[MC][i]+mu1*u[i-1]*f1[MC][i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
        #this is an auto regressive model, and the J(n) = f(n)^2

    #mu=0.1
    f2[MC][0]=u[0]
    w2[MC][0]=0
    mu2=0.1
    for i in range(N):
        f2[MC][i]=u[i]-w2[MC][i]*u[i-1]#predictor error
        w2[MC][i+1]=w2[MC][i]+mu2*u[i-1]*f2[MC][i]#LMS algorithm for adaptation of the (one and only) tap weight of the predictor
        #this is an auto regressive model, and the J(n) = f(n)^2

f_=np.zeros(N)
for i in range(N):
    for j in range(mc_sim):
        f_[i]+=f[j][i]   
    f_[i]=f_[i]/mc_sim

f1_=np.zeros(N)
for i in range(N):
    for j in range(mc_sim):
        f1_[i]+=f1[j][i]   
    f1_[i]=f1_[i]/mc_sim

f2_=np.zeros(N)
for i in range(N):
    for j in range(mc_sim):
        f2_[i]+=f2[j][i]   
    f2_[i]=f2_[i]/mc_sim

plt.figure(figsize=(20,10))
plt.plot(f_**2,'r')#learning curve, Mean square error vs number of adaptation cycles
plt.plot(f1_**2,'b')
plt.plot(f2_**2,'g')
plt.legend(('mu=0.01','mu=0.05','mu=0.2'),loc='upper right')
plt.xlabel('Number of adaptation cycles, n')
plt.ylabel('Mean square error')
plt.title('Learning curve')
plt.show()
```

The ensemble averaging was performed over 100 independent Monte Carlo
runs of the experiment. The plots of last chart confirm the following properties:
• As the step-size parameter m is reduced, the rate of convergence of the LMS algorithm
is correspondingly increased.
• A reduction in the step-size parameter m also has the effect of reducing variation
in the experimentally computed learning curve.
